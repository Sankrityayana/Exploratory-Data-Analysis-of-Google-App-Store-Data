```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import dataset using pandas
dfs = pd.read_csv("googleplaystore.csv")
```

```python
# Printing the forst 5 rows of our dataset
dfs.head()
```

```python
# Getting summary information about the dataset
dfs.info()
```

```python
# Descriptive statistics of dataset
dfs.describe()
```

```python
#Handle Missing values
dfs.dropna(inplace=True)        #drops the rows with missing values
```

```python
# Dealing with duplicates
dfs.drop_duplicates(inplace=True)       # drops duplicate rows
```

```python
dfs.info()
```

```python
# Changing the data type of Installs colum to numerical data type and remove the comma and plus symbols
dfs['Installs'] = dfs['Installs'].str.replace('+', '')
dfs['Installs'] = dfs['Installs'].str.replace(',', '')

dfs['Installs'] = dfs['Installs'].astype(int)

#Changing the data type of review column to numerical data type
dfs['Reviews'] = dfs['Reviews'].astype(int)
```

```python
dfs.head()
```

```python
dfs.info()
```

```python
# Bar plot
plt.figure(figsize=(10, 6))
sns.countplot(data=dfs, x='Category', hue='Category')
plt.xlabel('Category')
plt.xticks(rotation = 90)
plt.ylabel('Count')
plt.title('Distribution of App Categories')
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dfs, x='Reviews', y='Rating', hue='Category')
plt.xlabel('Number of Reviews')
plt.ylabel('Rating')
plt.title('Relationship between Reviews and Rating')
```

```python
# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=dfs, x='Category', y='Installs')
plt.xlabel('Category')
plt.xticks(rotation=90)
plt.ylabel('Installs')
plt.title('Distribution of Installs across Categories')
```

```python
dfs.head()
```

```python
from scipy.stats import ttest_ind

#splitting the dataset into free and paid apps
free_apps = dfs[dfs['Type'] == 'Free']
paid_apps = dfs[dfs['Type'] == 'Paid']
```

```python
# Performing t-test
t_stat, p_value = ttest_ind(free_apps['Rating'], paid_apps['Rating'])
```

```python
#Checking the p_value
if p_value < 0.005:
    print("Reject null hypothesis: There is a significant difference in ratings.")
else:
    print("Fail to reject null hypothesis: There is no significant difference in ratings.")
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=dfs, x='Rating', y='Installs', hue='Category')
plt.grid(True)
plt.xlabel('Rating')
plt.ylabel('Number of Installs')
plt.title('Relationship between Ratings and Installs')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
plt.show()
```

```python
# Line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=dfs, x='Last Updated', y='Rating')
plt.xlabel('Last Updated')
plt.xticks(rotation=90)
plt.ylabel('Average Rating')
plt.title('Average Rating Trend over Time')
```

```python
# Convert 'Last Updated' to datetime
dfs['Last Updated'] = pd.to_datetime(dfs['Last Updated'])

# Set 'Last Updated' as the index
dfs.set_index('Last Updated', inplace=True)

# Resample 'Rating' to monthly frequency
dfm = dfs['Rating'].resample('M').mean()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=dfm)
plt.xlabel('Date of Last Update (Monthly)')
plt.xticks(rotation=90)
plt.ylabel('Average Rating')
plt.title('Average Rating Trend over Time (Monthly)')
plt.show()
```
